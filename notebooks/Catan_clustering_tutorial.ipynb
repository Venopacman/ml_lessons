#### Правила Игры
Catan (или Колонизаторы) довольно известная настольная игра для 3-4 игроков в классическом издании. Для ясности хотелось бы пояснить некоторые интересные особенности правил игры:
    - Играют 3-4 игрока, цель игры - с помощью вырабатываемых ресурсов и торговли с соперниками набрать 10 очков развития быстрее остальных.
    
    - Поле состоит из гексоидов (шестиугольников), каждый гексоид отвечает за выработку одного из 5 ресурсов (дерево, глина, овца, камень, пшеница), исключением является гексоид-пустыня, он ничего не производит.
    
    - На гексоиде-пустыне находится фигурка разбойника,
    
    - На каждом гексоиде лежит фишка с числом от 2 до 12 (исключая 7), фишки раскладываются в случайном порядке, они отвечают за сбор того или иного ресурса.
    
    - Изначально на поле присутствует по 2 поселения для каждого игрока (итого 6 или 8 поселений на старте игры), поселения располагаются игроками на вершинах гексоидов по очереди, определяемой жребием, поселения и не могут распологаться с любым другим поселением на одном ребре гексоида.
    
    - После расстановки фишек и поселений, опредеяется очереднось хода игроков.
    
    - Каждый ход подчиняется единому ритуалу, который проходит в 3 части:
        - Сбор ресурсов - Игрок кидает 2 игральные кости, выпавшее число от 2 до 12 обозначает  гексоид, который приносит ресурс всем поселениям, расположенным на его вершинах.
        - Торговля - Игрок кидавший кость, может торговаться с любым соперником. Соперникам между собой торговаться запрещено.
        - Строительство - Игрок может потратить приобретенные ресурсы на постройку дорог, новых поселений, покупку бонусных карт или обновление поселения до города.      
    
    - Бросок костей с суммой равной 7 означает передвижение игроком разбойника с клетки пустыни на любой гексоид, тем самым блокируя сбор с гексоида ресурсов + все игроки обязаны скинуть половину из своих ресурсов, если у них накопилось больше 7 карт.

#### Описание данных
Для анализа и практики всевозможных алгоритмов нами был выбран дата-сет с 20-ю играми некоего юноши. Данные представлены в одной таблице, представленной следующими колонками:

- `L` - Дерево(lumber)

## Векторизация

Нужно что-то сделать с категориальными (номинальными) признаками, посмотрим есть ли среди категориальных признаков - бинарные:

```python
binary_columns    = [c for c in categorical_columns if data_describe[c]['unique'] == 2]
nonbinary_columns = [c for c in categorical_columns if data_describe[c]['unique'] > 2]
print("binary_columns: \n", binary_columns, "\n nonbinary_columns: \n", nonbinary_columns)
```

К небинарными признакам применим метод _векторизации_, 
который заключается в следующем.

Признак `j`, принимающий `s` значений, заменим на `s` признаков, принимащих значения `0` или `1`,
в зависимости от того, чему равно значение исходного признака `j`.

Например, в нашей задаче признак `settlement1_resourse_1_name` принимает 5 различных значений: 

```python
data['settlement1_resourse_1_name'].unique()
```

Заменим признак `settlement1_resourse_1_name` 5 признаками: `settlement1_resourse_1_name_L`, ..., `settlement1_resourse_1_name_C`.

 - Если признак `settlement1_resourse_1_name` принимает значение `L`, то признак `settlement1_resourse_1_name_L` равен `1`, а все остальные равны `0`.
 - Если признак `settlement1_resourse_1_name` принимает значение `C`, то признак `settlement1_resourse_1_name_C` равен `1`, а все остальные равны `0`.
 - и т.д.

Такую векторизацию осуществляет в pandas метод `get_dummies`:

```python
data_nonbinary = pd.get_dummies(data[nonbinary_columns])
print(data_nonbinary.columns)
```

## Нормализация количественных признаков

Многие алгоритмы машинного обучения чувствительны к масштабированию данных.

В этом случае количественные признаки полезно _нормализовать_.
Это можно делать разными способами.
Например, каждый количественный признак приведем к нулевому среднему и единичному среднеквадратичному отклонению:

Количество очков нормализовывать не надо, по ним будем классифицировать победителя.
Плюс, уберем поля идентифицирующие номер игры\порядковый номер пользователя\идентификатор нашего героя.

```python
from sklearn.model_selection import train_test_split
train_test_split_data = train_test_split(X, y, test_size = 0.2, random_state = 1)
X_train, X_test, y_train, y_test = train_test_split_data

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)
```

```python
def error_plot(classifier_name, parameter_name, parameter_values, train_errors, test_errors, cv_errors):
    plt.plot(parameter_values, train_errors, '.-', label='Train error')
    plt.plot(parameter_values, test_errors, '.-', label='Test error')
    plt.plot(parameter_values, cv_errors, '.-', label='CV error')
    plt.title(classifier_name)
    plt.xlabel(parameter_name)
    plt.ylabel('Error')
    plt.legend();


def print_errors(error_train, error_test):
    print("Train error: {0:.3f}\nTest error: {1:.3f}".format(error_train.item(), error_test.item()))

def get_errors(classifier, train_test_split_data):
    X_train, X_test, y_train, y_test = train_test_split_data
    
    y_train_predict = classifier.predict(X_train)
    y_test_predict = classifier.predict(X_test)
    
    error_train = np.mean(y_train != y_train_predict)
    error_test = np.mean(y_test != y_test_predict)
    return error_train, error_test
```

```python
def tune_model_parameter(classifier_name, classifier, parameter_name, parameter_values, train_test_split_data):
    X_train, X_test, y_train, y_test = train_test_split_data
    train_errors, test_errors, cv_errors = [], [], []
    
    for i in parameter_values:
#         print('{0}: {1}'.format(parameter_name, i))
        clf = classifier.set_params(**{parameter_name: i})
        clf.fit(X_train, y_train)
        
        train_error, test_error = get_errors(clf, train_test_split_data)
        train_errors.append(train_error)
        test_errors.append(test_error)
        
        cv = StratifiedShuffleSplit(n_splits=10, test_size=0.3, random_state=13)
        scores = cross_val_score(clf, X, y, cv=cv, scoring='accuracy')
        cv_errors.append(1 - scores.mean())
    
    min_cv_error = min(cv_errors)
    optimal_parameter_value = parameter_values[cv_errors.index(min_cv_error)]
    print('Min CV-error: {0:.4f} ({1}: {2:.4f})'.format(min_cv_error, parameter_name,  optimal_parameter_value))

    error_plot(classifier_name, parameter_name, parameter_values, train_errors, test_errors, cv_errors)
    
    return optimal_parameter_value
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

print_errors(*get_errors(knn, train_test_split_data))
```

```python
y_test_predict = knn.predict(X_test)
```

Получили 13 или 10 неверных ответов из 40. Можно отметить, что модель плохо определяет отрицательные таргеты - TNR << FPR. и отлично определяет положительные FNR = 0

Поиск К-ближайших соседей дал ошибку на тестовой выборке в 0.250, в добавок, модель недообучилась на обучающей выборке с ошибкой на тесте в 0.150.
Думаю это связано с банальностью признаков, в таких случаях стоит рассмотреть признаки выгоды изначального положения поселений (близость среднего по дайсу к 6 или 8)

# LDA, QDA

```python
import warnings
warnings.filterwarnings('ignore')
```

Посмотрим как покажет себя модель линейного дискриминантного анализа со стандарными параметрами для различных решателей

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
lda = LinearDiscriminantAnalysis(solver ='svd')
lda.fit(X_train, y_train)

print_errors(*get_errors(lda, train_test_split_data))
```

```python
y_test_predict = lda.predict(X_test)
print(confusion_matrix(y_test, y_test_predict))
```

```python
lda = LinearDiscriminantAnalysis(solver ='lsqr')
lda.fit(X_train, y_train)

print_errors(*get_errors(lda, train_test_split_data))
```

```python
y_test_predict = lda.predict(X_test)
print(confusion_matrix(y_test, y_test_predict))
```

Модель основанная на сингулярных разложениях, чуть больше обучилась, чем модель основанная на методе наименьших квадратов.
Однако обе модели показали идентичный результат на проверочной выборке. Заметим, что LDA также хуже распознает положительные таргеты, если сравнивать с KNN

Теперь посмотрим как покажет себя модель квадратичного дискриминантного анализа, и построим графики ошибок для различных параметров регрессии

```python
reg_param = np.arange(0.5, 1.5, 0.05)
best_reg_param = tune_model_parameter('QDA', QuadraticDiscriminantAnalysis(), 'reg_param', reg_param, train_test_split_data)
```

```python
qda = QuadraticDiscriminantAnalysis(reg_param=best_reg_param)
qda.fit(X_train, y_train)

print_errors(*get_errors(qda, train_test_split_data))
```

```python
y_test_predict = qda.predict(X_test)
print(confusion_matrix(y_test, y_test_predict))
```

```python
qda = QuadraticDiscriminantAnalysis(reg_param=0.9)
qda.fit(X_train, y_train)

print_errors(*get_errors(qda, train_test_split_data))
```

```python
y_test_predict = qda.predict(X_test)
print(confusion_matrix(y_test, y_test_predict))
```

Опять видим, что значение средней ошибки в методе скользящей оценки (кросс-валидации) не обязательно соотвествует минимальному значению ошибки на проверочной выборке. Заметим, что QDA с регрессионным параметром = 0.9 показывает более равномерную способность к распознованию как положительных так и отрицательных таргетов, вкупе с неплохой ошибкой на проверочной выборке = 0.2

Получается, что на данным момент лучше свего себя показывает QDA
