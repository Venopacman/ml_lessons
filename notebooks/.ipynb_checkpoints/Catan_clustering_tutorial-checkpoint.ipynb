#### Правила Игры
Catan (или Колонизаторы) довольно известная настольная игра для 3-4 игроков в классическом издании. Для ясности хотелось бы пояснить некоторые интересные особенности правил игры:
    - Играют 3-4 игрока, цель игры - с помощью вырабатываемых ресурсов и торговли с соперниками набрать 10 очков развития быстрее остальных.
    
    - Поле состоит из гексоидов (шестиугольников), каждый гексоид отвечает за выработку одного из 5 ресурсов (дерево, глина, овца, камень, пшеница), исключением является гексоид-пустыня, он ничего не производит.
    
    - На гексоиде-пустыне находится фигурка разбойника,
    
    - На каждом гексоиде лежит фишка с числом от 2 до 12 (исключая 7), фишки раскладываются в случайном порядке, они отвечают за сбор того или иного ресурса.
    
    - Изначально на поле присутствует по 2 поселения для каждого игрока (итого 6 или 8 поселений на старте игры), поселения располагаются игроками на вершинах гексоидов по очереди, определяемой жребием, поселения и не могут распологаться с любым другим поселением на одном ребре гексоида.
    
    - После расстановки фишек и поселений, опредеяется очереднось хода игроков.
    
    - Каждый ход подчиняется единому ритуалу, который проходит в 3 части:
        - Сбор ресурсов - Игрок кидает 2 игральные кости, выпавшее число от 2 до 12 обозначает  гексоид, который приносит ресурс всем поселениям, расположенным на его вершинах.
        - Торговля - Игрок кидавший кость, может торговаться с любым соперником. Соперникам между собой торговаться запрещено.
        - Строительство - Игрок может потратить приобретенные ресурсы на постройку дорог, новых поселений, покупку бонусных карт или обновление поселения до города.      
    
    - Бросок костей с суммой равной 7 означает передвижение игроком разбойника с клетки пустыни на любой гексоид, тем самым блокируя сбор с гексоида ресурсов + все игроки обязаны скинуть половину из своих ресурсов, если у них накопилось больше 7 карт.

#### Описание данных
Для анализа и практики всевозможных алгоритмов нами был выбран дата-сет с 20-ю играми некоего юноши. Данные представлены в одной таблице, представленной следующими колонками:

- `gameNum` - Номер игры, каждая игра проводилась для 4х игроков, следовательно на игру в таблице будет уделено 4 строчки под каждого игрока.

- `player` - Начальная позиция

- `points` - С каким количеством очков игрок окончил игру (игра заканчивается, когда игрок набирает не меньше 10 очков).

- `me` - Пометка, свидетельствующая о том, что данным игроком был автор дата-сета.

- `2`, `3`, ..., `12` - Количество выпадений определенной суммы на 2-х игральных костях в игре для игрока.

- `settlement1`, `settlement2` - Характеристики начальных поселений игрока, каждое поселение описано 3-мя парами [номер присвоенный гексоиду, ресурс]:

- `L` - Дерево(lumber)

- `C` - Глина(clay)

- `S` - Овца(sheep)

- `W` - Пшеница(wheat)

- `O` - Камень(ore)

- `3G` - 3:1 общий порт (Доступ к порту, позволяет обменивать 3 единицы одного ресурса на любой с "мировым рынком")

- `2(X)` - 2:1 Порт ресурса-`X` (Улучшенные возмржности общего порта с обменом два к одному, только для строго обозначеннного ресурса)

- `D` - пустыня

`Пример`: В первой игре, первое поселение игрока 1 находится на вершинах гексоидов с фишкой с номером 6 и деревом, фишкой с номером 3 и глиной  и фишкой с номером 11 и глиной.

- `production` - общее количество ресурсов в шт. произведенное поселениеями и городами в течение игры.

- `tradeGain`- общее количество ресурсов в шт. заработанное обменом с соперниками и "мировым рынком" в течение игры.

- `robberCardsGain` - общее количество ресурсов в шт. украденных с помощью разбойника + карты развития без учета рыцарей в течение игры. Карта строительства дорог, дает +4 к признаку(т.к. строительство дороги эквивалентно 2 шт. ресурсов).

- `totalGain` - сумма предыдущих 3-х колонок.

- `tradeLoss` - количество отданных в ходе обмена с соперниками и "мировым рынком" карт в течение игры.

- `robberCardsLoss` - общее количество кард отданных под воздействием разбойника или карты развития типа монополия. 

- `tribute` - общее количество карт, потерянных при выпадении на костях суммы равной 7 (отделено от предыдущих колонок потерь)

- `totalLoss` - сумма предыдущих 3-х колонок.

- `totalAvailable` - totalGain минус totalLoss.

## Векторизация

Нужно что-то сделать с категориальными (номинальными) признаками, посмотрим есть ли среди категориальных признаков - бинарные:

```python
binary_columns    = [c for c in categorical_columns if data_describe[c]['unique'] == 2]
nonbinary_columns = [c for c in categorical_columns if data_describe[c]['unique'] > 2]
print("binary_columns: \n", binary_columns, "\n nonbinary_columns: \n", nonbinary_columns)
```

К небинарными признакам применим метод _векторизации_, 
который заключается в следующем.

Признак `j`, принимающий `s` значений, заменим на `s` признаков, принимащих значения `0` или `1`,
в зависимости от того, чему равно значение исходного признака `j`.

Например, в нашей задаче признак `settlement1_resourse_1_name` принимает 5 различных значений: 

```python
data['settlement1_resourse_1_name'].unique()
```

Заменим признак `settlement1_resourse_1_name` 5 признаками: `settlement1_resourse_1_name_L`, ..., `settlement1_resourse_1_name_C`.

 - Если признак `settlement1_resourse_1_name` принимает значение `L`, то признак `settlement1_resourse_1_name_L` равен `1`, а все остальные равны `0`.
 - Если признак `settlement1_resourse_1_name` принимает значение `C`, то признак `settlement1_resourse_1_name_C` равен `1`, а все остальные равны `0`.
 - и т.д.

Такую векторизацию осуществляет в pandas метод `get_dummies`:

```python
data_nonbinary = pd.get_dummies(data[nonbinary_columns])
print(data_nonbinary.columns)
```

## Нормализация количественных признаков

Многие алгоритмы машинного обучения чувствительны к масштабированию данных.
К таким алгоритмам, например, относится метод ближайших соседей, машина опорных векторов и др.

В этом случае количественные признаки полезно _нормализовать_.
Это можно делать разными способами.
Например, каждый количественный признак приведем к нулевому среднему и единичному среднеквадратичному отклонению:

Количество очков нормализовывать не надо, по ним будем классифицировать победителя.
Плюс, уберем поля идентифицирующие номер игры\порядковый номер пользователя\идентификатор нашего героя.

Разобъем наш дата-сет на 2 выборки обучающую и проверочную, проверочная выборка необходима для оценки качества обучения модели. Так как объм обучающих данных оставляет желать лучшего, возьмем пропорцию разбиения 4:1 = 20% на проверочую выборку и 80% на обучение

```python
from sklearn.model_selection import train_test_split
train_test_split_data = train_test_split(X, y, test_size = 0.2, random_state = 1)
X_train, X_test, y_train, y_test = train_test_split_data

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)
```

```python
def error_plot(classifier_name, parameter_name, parameter_values, train_errors, test_errors, cv_errors):
    plt.plot(parameter_values, train_errors, '.-', label='Train error')
    plt.plot(parameter_values, test_errors, '.-', label='Test error')
    plt.plot(parameter_values, cv_errors, '.-', label='CV error')
    plt.title(classifier_name)
    plt.xlabel(parameter_name)
    plt.ylabel('Error')
    plt.legend();


def print_errors(error_train, error_test):
    print("Train error: {0:.3f}\nTest error: {1:.3f}".format(error_train.item(), error_test.item()))

def get_errors(classifier, train_test_split_data):
    X_train, X_test, y_train, y_test = train_test_split_data
    
    y_train_predict = classifier.predict(X_train)
    y_test_predict = classifier.predict(X_test)
    
    error_train = np.mean(y_train != y_train_predict)
    error_test = np.mean(y_test != y_test_predict)
    return error_train, error_test
```

```python
def tune_model_parameter(classifier_name, classifier, parameter_name, parameter_values, train_test_split_data):
    X_train, X_test, y_train, y_test = train_test_split_data
    train_errors, test_errors, cv_errors = [], [], []
    
    for i in parameter_values:
#         print('{0}: {1}'.format(parameter_name, i))
        clf = classifier.set_params(**{parameter_name: i})
        clf.fit(X_train, y_train)
        
        train_error, test_error = get_errors(clf, train_test_split_data)
        train_errors.append(train_error)
        test_errors.append(test_error)
        
        cv = StratifiedShuffleSplit(n_splits=10, test_size=0.3, random_state=13)
        scores = cross_val_score(clf, X, y, cv=cv, scoring='accuracy')
        cv_errors.append(1 - scores.mean())
    
    min_cv_error = min(cv_errors)
    optimal_parameter_value = parameter_values[cv_errors.index(min_cv_error)]
    print('Min CV-error: {0:.4f} ({1}: {2:.4f})'.format(min_cv_error, parameter_name,  optimal_parameter_value))

    error_plot(classifier_name, parameter_name, parameter_values, train_errors, test_errors, cv_errors)
    
    return optimal_parameter_value
```

# $k$NN

```python
from math import sqrt
n_neighbors = range(1, int(sqrt(X_train.shape[0])))
```

```python
from sklearn.neighbors import KNeighborsClassifier
best_k = tune_model_parameter('KNN', KNeighborsClassifier(), 'n_neighbors', n_neighbors, train_test_split_data)
```

```python
knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(X_train, y_train)

print_errors(*get_errors(knn, train_test_split_data))
```

```python
y_test_predict = knn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_test_predict))
```

Можно отметить, что хоть метод скользящей оценки дает минимальную ошибку для k=6, 
мы можем наблюдать ошибку на проверочной выборке с k=5 меньше "оптимальной" как на обучающей, так и на проверочной выборке.

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

print_errors(*get_errors(knn, train_test_split_data))
```

```python
y_test_predict = knn.predict(X_test)
```

Получили 13 или 10 неверных ответов из 40. Можно отметить, что модель плохо определяет отрицательные таргеты - TNR << FPR. и отлично определяет положительные FNR = 0

Поиск К-ближайших соседей дал ошибку на тестовой выборке в 0.250, в добавок, модель недообучилась на обучающей выборке с ошибкой на тесте в 0.150.
Думаю это связано с банальностью признаков, в таких случаях стоит рассмотреть признаки выгоды изначального положения поселений (близость среднего по дайсу к 6 или 8)

```python
rf = ensemble.RandomForestClassifier(n_estimators=100, oob_score=True,random_state=123)
rf.fit(X_train, y_train)

print_errors(*get_errors(rf, train_test_split_data))
```

```python
n_estimators = range(1,1001,50)
```

Построим зависимость ошибок от количества решающих деревьев при прочих равных

```python
test_errors = []
train_errors = []
oob_errors = []
for n_trees in n_estimators:
    rf = ensemble.RandomForestClassifier(n_estimators=n_trees, oob_score=True,random_state=123)
    rf.fit(X_train, y_train)
    train_errors.append(1 - sum(rf.predict(X_train) == y_train)/len(y_train))
    test_errors.append(1 - sum(rf.predict(X_test) == y_test)/len(y_test))
    oob_errors.append(1 - rf.oob_score_)

    
plt.plot(n_estimators, test_errors,  'b-o', label = 'test error')
plt.plot(n_estimators, train_errors,  'r-o', label = 'train error')
plt.plot(n_estimators, oob_errors, 'g-o', label = 'oob error')
plt.title('Error vs. number of trees')
plt.xlabel('Trees')
plt.ylabel('Error')
plt.legend()
pass
```

Найдем оптимальное количество деревьев для нашей модели, оно будет соответсвовать минимальной ошибке out-of-box

# Principal component analysis (PCA)

Уменьшим размерность данных с помощью метода главных компонент и посмотрим, поможет ли это нам улучшить нашу 'чемпионскую модель' AdaBoost c параметрами
- ensemble.AdaBoostClassifier(n_estimators = 11, random_state=123, learning_rate=1.8368163265306121)

```python
from sklearn.decomposition import PCA
```

```python
decomposition_estimate = range(1, X_train.shape[1] +1)
```

```python
test_errors, train_errors, cv_errors = [],[],[]
for decomp_num in decomposition_estimate:
    pca = PCA(n_components = decomp_num)
    pca.fit(np.concatenate((X_train,X_test)))
    X_train_reduc = pca.transform(X_train)
    X_test_reduc = pca.transform(X_test)
    
    ada = ensemble.AdaBoostClassifier(n_estimators = 11, random_state=123, learning_rate=best_lr)
    ada.fit(X_train_reduc, y_train)

    train_error = np.mean(y_train != ada.predict(X_train_reduc))
    test_error = np.mean(y_test  != ada.predict(X_test_reduc))
    train_errors.append(train_error)
    test_errors.append(test_error)
    X = np.concatenate((X_test_reduc, X_train_reduc))
    y = np.concatenate((y_test,y_train))
    cv = StratifiedShuffleSplit(n_splits=10, test_size=0.3, random_state=123)
    scores = cross_val_score(ada, X, y, cv=cv, scoring='accuracy')
    cv_errors.append(1 - scores.mean())
    
plt.plot(decomposition_estimate, test_errors,  'b-o', label = 'test error')
plt.plot(decomposition_estimate, train_errors,  'r-o', label = 'train error')
plt.plot(decomposition_estimate, cv_errors,  'g-o', label = 'cv error')
plt.title('Error vs. num of components')
plt.xlabel('# of components')
plt.ylabel('Error')
plt.legend()
pass
```

```python
best_decomp_num = decomposition_estimate[np.argmin(cv_errors)]
best_decomp_num
```

```python
pca = PCA(n_components = best_decomp_num)
pca.fit(np.concatenate((X_train,X_test)))
X_train_reduc = pca.transform(X_train)
X_test_reduc = pca.transform(X_test)

ada = ensemble.AdaBoostClassifier(n_estimators = 11, random_state=123, learning_rate=best_lr)
ada.fit(X_train_reduc, y_train)

train_error = np.mean(y_train != ada.predict(X_train_reduc))
test_error = np.mean(y_test  != ada.predict(X_test_reduc))
print(train_error, test_error)
```

Метод главных компонент, не только не улучшил качество нашей модели, но даже умудрился испортить нашу чемпионскую точность (приведена ниже, для успокоения нервов), причем в значительной мере и даже для случая преобразования в пространство размерности эквивалентное оригинальному (70-ти мерное)

```python
ada = ensemble.AdaBoostClassifier(n_estimators = 11, random_state=123, learning_rate=best_lr)
ada.fit(X_train, y_train)

train_error = np.mean(y_train != ada.predict(X_train))
test_error = np.mean(y_test  != ada.predict(X_test))
print(train_error, test_error)
```

Подводя итог, можно сказать, что модель адаптивного бустинга AdaBoost дала наилуучшую точность, достигнутую благодаря настройке скорости обучения и качественного выбора количества классификаторов в модели. Хорошие результаты наблюдались у алгоритмов QDA и GradientBoostingTrees, но они имели менее сильную обобщающую способность. Метод главных компонент, к сожалению, не дал положительного эффекта к наилучшей модели, наоборот, сильно ухудшив точность и усложнив обучение на прежних параметрах.
